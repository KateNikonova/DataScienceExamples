## Предобработка данных

Посчитаем сколько пропусков в таблице для каждого столбца.

```python
df.isna().sum()
```

Рассмотрим все пропуски по очереди.

### Обработка столбца с количеством дней размещения объявления

Начнем обработку со столбца *days_exposition* - количество дней, которые было размещено объявление (от публикации до снятия). Пропуск в этой колонке значит что объявление еще не снято с публикации. Таких объявлений 3181 из 23699. Пропуски можно оставить, но для вычислений лучше тип в этой колонке привести к float.

```python
df['days_exposition'] = df['days_exposition'].astype('float')
df['days_exposition'] = df['days_exposition'].apply(lambda x: f"{x:.2f}")
df['days_exposition'] = df['days_exposition'].astype('float')

# посмотрим общее инфо по столбцу
df['days_exposition'].describe()
```

### Обработка столбцов локация, наличие парков и водоемов и расстояние до них

Далее рассмотрим пропуски для группы parks_around3000, parks_nearest, ponds_around3000, ponds_nearest.
Логично предположить что, если значение в parks_around3000 и ponds_around3000 пустое, то парков и водоемов нет и можно заполнить 0. Но нужно проверить, пустые ли при этом parks_nearest и ponds_nearest.

```python
parks_nan = df.loc[df['parks_around3000'].isna()]
parks_nan['parks_nearest'].describe()
```

```python
ponds_nan = df.loc[df['ponds_around3000'].isna()]
ponds_nan['ponds_nearest'].describe()
```

```python
display(len(parks_nan))
display(len(ponds_nan))
```

Все пустые значения в столбцах parks_around3000 и ponds_around300 можно заполнить 0 и после привести типу к int64. Но так как в столбцах значения с плавающей точкой, то сначала приведем к float, а потом к int

```python
df['parks_around3000'] = df['parks_around3000'].fillna(value=0)
df['ponds_around3000'] = df['ponds_around3000'].fillna(value=0)
df = df.astype({'parks_around3000': float, 'ponds_around3000': float})
df = df.astype({'parks_around3000': int, 'ponds_around3000': int})
```

Теперь заполним пропуски в parks_nearest и ponds_nearest.
Если значения в столбцах parks_around3000 и ponds_around300 - 0, то parks_nearest и ponds_nearest можно так же оставить пустыми. А для остальных логичнее указать среднее значение по локации.

```python
# приведем типы столбцов к числовому 
df = df.astype({'parks_nearest': float, 'ponds_nearest': float})

# исключим строки с пустой локацией, их всего 49 из 23699
df = df.query('locality_name.notna()')
```

```python
#расчитаем среднее по локации для parks_nearest и ponds_nearest и присвоим эти значения пустым 
park_grouped = df.groupby('locality_name')['parks_nearest'].mean()
 
def park_func(row):
    if pd.isna(row['parks_nearest']) and row['parks_around3000'] > 0:
        return park_grouped.loc[row['locality_name']]
    return row['parks_nearest']
 
df['parks_nearest'] = df.apply(park_func, axis=1)
```

```python
pond_grouped = df.groupby('locality_name')['ponds_nearest'].mean()
 
def pond_func(row):
    if pd.isna(row['ponds_nearest']) and row['ponds_around3000'] > 0:
        return park_grouped.loc[row['locality_name']]
    return row['ponds_nearest']
 
df['ponds_nearest'] = df.apply(pond_func, axis=1)
```

```python
#отредактируем значения в столбцах, оставив 2 знака после запятой. 
df['parks_nearest'] = df['parks_nearest'].apply(lambda x: f"{x:.2f}")
df['ponds_nearest'] = df['ponds_nearest'].apply(lambda x: f"{x:.2f}")
df = df.astype({'parks_nearest': float, 'ponds_nearest': float})
```

Проверим что не осталось пустых значений для parks_nearest и ponds_nearest для строк с ненулевыми значениями в столбцах parks_around3000 и ponds_around300.

```python
display(len(df.query('parks_around3000 > 0 and parks_nearest.isna()')))
display(len(df.query('ponds_around3000 > 0 and ponds_nearest.isna()')))
```

### Обработка столбцов airports_nearest и cityCenters_nearest

Поменяем тип данных на float для столбцов airports_nearest и cityCenters_nearest

```python
df = df.astype({'airports_nearest': float, 'cityCenters_nearest': float})
display(len(df.query('airports_nearest.isna()')))
display(len(df.query('cityCenters_nearest.isna()')))
```

Пропуски можно заполнить средним значением по локации. И проверить останутся ли пропуски.

```python
#найдем среднее значение для расстояния до аэропорта по локации и заполним пропуски
airports_grouped = df.groupby('locality_name')['airports_nearest'].mean()
 
def air_func(row):
    if pd.isna(row['airports_nearest']):
        return airports_grouped.loc[row['locality_name']]
    return row['airports_nearest']
 
df['airports_nearest'] = df.apply(air_func, axis=1)


#найдем среднее значение для расстояния до центра по локации и заполним пропуски
center_grouped = df.groupby('locality_name')['cityCenters_nearest'].mean()
 
def center_func(row):
    if pd.isna(row['cityCenters_nearest']):
        return center_grouped.loc[row['locality_name']]
    return row['cityCenters_nearest']
 
df['cityCenters_nearest'] = df.apply(center_func, axis=1)
```

```python
#отредактируем значения в столбцах, оставив 2 знака после запятой. 
df['airports_nearest'] = df['airports_nearest'].apply(lambda x: f"{x:.2f}")
df['cityCenters_nearest'] = df['cityCenters_nearest'].apply(lambda x: f"{x:.2f}")
df = df.astype({'airports_nearest': float, 'cityCenters_nearest': float})

#проверим оставшееся количество пропусков
display(len(df.query('airports_nearest.isna()')))
display(len(df.query('cityCenters_nearest.isna()')))
```

Пропуски все равно остались, и это нормально. Значит среднее значение по локации тоже пусто.

### Обработка столбцов kitchen_area  и living_area

- kitchen_area как разницу total_area и living_area
- living_area как разницу total_area и kitchen_area

Сгруппируем данные по типу планировки и посмотрим средние значения для kitchen_area и leaving_area.

```python
kitch_grouped = df.groupby('open_plan')['kitchen_area'].mean()
liv_grouped = df.groupby('open_plan')['living_area'].mean()
display(kitch_grouped)
display(liv_grouped)
```

Жилую площадь можно заполнить средним значениями, для кейсов когда и кухня и жилая площадь имеют значения пустые. Площадь кухни для недвижимости с открытой планировкой среднее значение NaN. Это скорее всего значит что кухни нет. Поэтому такие пропуски заполним 0.

```python
kitch_grouped[1] = 0
display(kitch_grouped)
```

```python
def kitch_func(row):
    if pd.isna(row['kitchen_area']) and pd.isna(row['living_area']):
        return kitch_grouped.loc[row['open_plan']]
    return row['kitchen_area']
 
df['kitchen_area'] = df.apply(kitch_func, axis = 1)
```

```python
def liv_func(row):
    if pd.isna(row['living_area']) and pd.isna(row['kitchen_area']):
        return liv_grouped.loc[row['open_plan']]
    return row['living_area']

df['living_area'] = df.apply(liv_func, axis = 1)
```
